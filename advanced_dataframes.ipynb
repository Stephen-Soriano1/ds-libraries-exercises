Copy the users and roles DataFrames from the examples above.

```python
users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users
```

```python
roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
```

What is the result of using a right join on the DataFrames?


```python
df = pd.merge(users, roles, how='right', left_on='role_id', right_on='id')
df
```

```python
df_concat = pd.concat([users,roles], axis=1)
df_concat
```

What is the result of using an outer join on the DataFrames?


```python
df_out = pd.merge(users, roles, how='outer', left_on='role_id', right_on='id')
df_out
```

What happens if you drop the foreign keys from the DataFrames and try to merge them?


```python
df_mess = pd.merge(users, roles, how='right', left_on='roleid', right_on='id')
df_mess
```

Create a column named mileage_difference like you did in the DataFrames exercises; this column should contain the difference between highway and city mileage for each car.


```python
cars['mileage_difference'] = cars.hwy - cars.cty
cars
```

Create a column named average_mileage like you did in the DataFrames exercises; this is the mean of the city and highway mileage.


```python
cars['average_mileage'] = cars[['cty','hwy']].mean(axis=1)
cars
```

Create a new column on the mpg dataset named is_automatic that holds boolean values denoting whether the car has an automatic transmission.

```python
cars.trans.sort_values()
```

```python
cars['is_automatic'] = cars['trans'].str.contains('auto')
cars
```

Using the mpg dataset, find out which which manufacturer has the best miles per gallon on average?

```python
cars[['manufacturer','average_mileage']].groupby('manufacturer').max()
```

Do automatic or manual cars have better miles per gallon?


```python
cars.groupby('is_automatic').mean().average_mileage
```

Write the code necessary to create a cross tabulation 
of the number of titles by department. 
(Hint: this will involve a combination of SQL code to pull the necessary data and 
 python/pandas code to perform the manipulations.)

```python
query = '''
    select *
    from dept_emp
        join departments
            using (dept_no)
'''
```
