Load the chipotle dataset from SQL, create a bar chart that shows the 4 most popular items and the revenue produced by each.


```python
def get_db_url(user, password, host, db):
    return (f'mysql+pymysql://{user}:{password}@{host}/{db}')
```

```python
url = get_db_url(eny.user, eny.password, eny.host, 'chipotle')
```

```python
query = 'select * from orders'
```

```python
df = pd.read_sql(query, url)
```

```python
df.head()
```

```python
df.dtypes
```

```python
items_df = df.groupby('item_name')[['quantity','item_price']].sum()
items_df.head()
```

```python
top_items = items_df.sort_values('quantity', ascending=False).head(4)
top_items
```

```python
top_items.reset_index()
```

```python
sns.barplot(data=top_items, y=top_items.index, x='item_price');
```
